```python
import chromadb
from openai import OpenAI
from pypdf import PdfReader
from pathlib import Path  # Add this line
```

```python
# Initialize persistent ChromaDB
client = chromadb.PersistentClient(path="./policy_db")
# Create collection with sentence-transformers embedding
collection = client.get_or_create_collection(
    name="policy_documents",
    metadata={"description": "Company policy documents"}
)
```

```python
# Load all PDFs
policies_path = Path("./policies")

documents = []
metadatas = []
ids = []

for pdf_file in policies_path.glob("*.pdf"):
    print(f"Processing: {pdf_file.name}")
    
    reader = PdfReader(pdf_file)
    text = " ".join([page.extract_text() or "" for page in reader.pages])
    
    documents.append(text)
    metadatas.append({"filename": pdf_file.name, "doc_type": "policy"})
    ids.append(pdf_file.stem)  # filename without extension

print(f"\nLoaded {len(documents)} documents")
```
